```python
#define the matrices A, B and define vectors x, y
A = matrix([[1,2],[0, 1],[2, 3]])
B = matrix([[1,3,2],[2,5,1],[4,1,0]])
x = vector(SR,var('x1 x2'))
y = vector([25, 15])

print("The matrix A:")
print(A)

print("\nThe matrix B:")
print(B)

print("\nVector x:")
print(x)

print("\nVector y:")
print(y)
```

```python
#compute BB^T and check if it is singular
BBT = B*B.T
print("Matrix BB^T is:")
print(B*B.T)

detBBT = BBT.det
if detBBT == 0:
    print("\nBBT is singular")
else:
    print("\nBBT is not singular")
```

```python
#compute C=A.T*(BB.T)^(-1)*A
C = A.T*(B*B.T).inverse()*A
C
```

```python
#solve the inear system Cx=y in terms of x=(x1, x2)
##using matrix inversion
sol_inv = C.inverse()*y
print("Cx=y with matrix inversion is:")
print(sol_inv)

##using solve function
x1, x2 = var('x x2')
eqs = list(C*vector([x1, x2])-y)
sol_solve = solve(eqs, [x1, x2])
print("\nCx=y with solve function is:")
sol_solve
```
