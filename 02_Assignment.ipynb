Problem 7


```sagemath
#vector v
v = vector([3,2])
```

```sagemath
#matrix that rotate the vector counterclokwise by 15 degrees (15 = 180/12 = π/12)
a=pi/12 
A = matrix(SR,[[cos(a), -sin(a)],
               [sin(a),  cos(a)]])
A
```

```sagemath
#matrix that rotate the vector counterclokwise by 45 degrees (45 = π/4)
b=pi/4
B = matrix(SR,[[cos(b), -sin(b)],
               [sin(b),  cos(b)]])
B
```

```sagemath
#compute the rotated vector vA
vA = A*v
vA
```

```sagemath
#compute the rotated vector vB
vB = B*v
vB
```

```sagemath
#sketch vectors vA, vB and v on a coordinate grid
origin = vector([0,0])
Arrows = arrow(origin, v, color='red', width=2, legend_label='v') +\
arrow(origin, vA, color='blue', width=2, legend_label='vA') +\
arrow(origin, vB, color='green', width=2, legend_label='vB')
Arrows.show(aspect_ratio=1)
```

```sagemath
#compare the lenght of vectors vA, vB and v
Vnorm = v.norm().simplify_full()
VAnorm = vA.norm().simplify_full()
VBnorm = vB.norm().simplify_full()
Vnorm, VAnorm, VBnorm
```

Comment:

Matrices A and B are rotation matrices. Any rotation matrix is orthogonal and preserve the length of vectors. That's why the lengths of v, vA, and vB are equal.



```sagemath
#find the angle (in degrees) between vA and vB
angle = acos((vA.dot_product(vB)) / (vA.norm()*vB.norm()))
round(float(angle*180/pi))
```

```sagemath
#vectors
u1 = vector([1,0,2,1,3])
u2 = vector([0,1,1,2,1])
u3 = vector([2,1,0,1,4])
```

```sagemath
#matrix A=[u1,u2,u3]
A = column_matrix([u1,u2,u3])
A
```

```sagemath
#check whether the vectors u1, u2, u3 are linearly independent
if A.rank() == 3:
    print ('Vectors v1,v2 and v3 are linearly independent')
else:
    print ('Vectors are linearly dependent')
```

```sagemath
#find the projection matrix P that maps R^5 onto U
P = A * (A.transpose()* A).inverse() * A.transpose()
P
```

```sagemath
#find the projection of the vector x onto U
x = vector([3,-1,4,0,2])
Px = P * x
Px
```

```sagemath
#compute the projection error e=x-Px
e = x - Px
e
```

```sagemath
#verify that e is orthogonal to U
if (e.dot_product(u1), e.dot_product(u2), e.dot_product(u3)) == (0,0,0):
    print('e is orthogonal to U')
else:
    print('e is not orthogonal to U')
(e.dot_product(u1), e.dot_product(u2), e.dot_product(u3))
```

```sagemath
#compute norm of e
enorm = e.norm()
enorm
```
